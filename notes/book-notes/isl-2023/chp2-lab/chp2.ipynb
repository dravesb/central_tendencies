```python
y = x + np.random.normal(loc=50, scale=1, size=50)
```

```python
# correlation coefficient
np.corrcoef(x, y)
```

```python
# we have different seeds so
print(np.random.normal(scale=5, size=2))
print(np.random.normal(scale=5, size=2))
```

```python
# we can set a random seed by
rng = np.random.default_rng(1303)
print(rng.normal(scale=5, size=2))
rng2 = np.random.default_rng(1303)
print(rng2.normal(scale=5, size=2))
```

```python
# we can use this to show how to get empirical moments

rng = np.random.default_rng(1234)
y = rng.standard_normal(10)
np.mean(y), y.mean()
```

```python
# variance - notice these are not biased corrected
np.var(y), y.var(), np.mean((y - y.mean()) ** 2)
```

```python
# we can change this by setting the ddof argument
np.var(y, ddof=1), y.var(ddof=1)
```

```python
np.sqrt(np.var(y)), y.std()
```
