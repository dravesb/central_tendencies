```python
A = np.array(np.arange(16)).reshape((4, 4))
A
```

```python
A[1, 2]
```

```python
# we can pass a list to get a subset of rows
row_lst = [1, 3]
A[row_lst]
```

```python
# we can also access columns this way by passing it an explicit : rows
col_lst = [0, 2]
A[:, col_lst]
```

```python
# submatries are based on intersections
A[row_lst, col_lst]
```

```python
# to manually do this we have to first get the rows, then the columns
A[row_lst, :][:, col_lst]
```

```python
# We can do this more efficiently via an intermediate mask
idx = np.ix_([1, 3], [0, 2, 3])
```

```python
idx
```

```python
A[idx]
```

```python
# we can also use slice notation
A[1:4:2, 0:3:2]
```

### Boolean Indexing

```python
keep_rows = np.zeros(A.shape[0], bool)
keep_rows
```

```python
keep_rows[[1, 3]] = True
keep_rows
```

```python
np.all(keep_rows == np.array([0, 1, 0, 1]))
```

```python
# np.all checks whether all entries are true
# np.any checks wehter any entry is true
```

```python
A[np.array([0, 1, 0, 1])]
```

```python
# keep rows applys boolean indexing the length of the arrary
A[keep_rows]
```

```python
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]
```

```python
idx_mixed = np.ix_([1, 3], keep_cols)
A[idx_mixed]
```

### For loops

```python
total = 0
for value in [3, 2, 19]:
    total += value
print(f"Total value is: {total}")
```

```python
total = 0
for value in [2, 3, 19]:
    for weight in [3, 2, 1]:
        total += value * weight

print(f"Total is: {total}")
```

```python
total = 0
for v, w in zip([2, 3, 19], [0.2, 0.3, 0.5]):
    total += v * w

print(f"Weighted average is {total}")
```
